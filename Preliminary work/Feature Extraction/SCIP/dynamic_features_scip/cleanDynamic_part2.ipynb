```python
# Now make a reusable function out of this

def extract_table(df, start, end, tablename, tablesize):
    for column in df.columns[start+1:end]:
        #print(df[column])
        df_new = df[column].str.split(expand=True)
        name_list = []
        for name in df[tablename][0].split():
            name_list.append(name)
        #print(df_new)
        for i in range(tablesize):
            df_new.rename(columns={i: tablename + '_' + column + '_' + name_list[i]}, inplace=True)
        #print(df_new)
        #print('########')
        df = pd.merge(df, df_new, left_index=True, right_index=True)

    for i in df.columns[start:end]:
        df.drop(i, inplace=True, axis=1)

    return df

#example_table = pd.DataFrame([['as 1','bs 2', 'ts 3','vs 4','ps 5'],['rs 5','ys 4','io 3','oo 2','qw 1'],['ee 9','dd 8','ui 7','we 6','as 5'],['tf 5','zx 5','yf 5','zx 5','yt 5'],['qw 1','ut 1','fg 1','jk 1',' vb 1']], columns=['A','B','C','D','E'])
#example_table.head()
#extract_table(example_table, start=2, end=4, tablename='C', tablesize=2)
```

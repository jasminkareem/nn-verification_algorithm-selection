## Making a .csv from SCIP's feature extractor
In this notebook, we already have the extracted features from scip's fetuare extractor. These files are stored as .txt files, each representing the features of a particular instance. So we just need to turn this into a dataframe that we can use to make predictions

```python
directory = 'C:/Users/Jasmin/Documents/MSc-Thesis/Data/ExtractedFeatures/extracted_features_presolve_off_scip_10k/'

all_mips = {}
file_order = []

for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(f):
        print("*****", filename, "*****")
        file_order.append(filename)
        with open(f, "r") as filestream:
            for line in filestream:
                currentline = line.split(",")
                all_mips[str(filename)] = currentline
```

```python
with open("C:/Users/Jasmin/Documents/MSc-Thesis/Feature_Extraction/feature_extractor/src/features_description.csv") as f:
    feature_description = [row.split(',')[0] for row in f]

feature_description    
```

```python
len(all_mips['mip_93.txt'])
```

```python
all_mips['feature_name'] = feature_description
```

```python
feature_df = pd.DataFrame.from_dict(all_mips, orient='index', columns=all_mips['feature_name'])
```

```python
feature_df = feature_df.drop(['feature_name'])

feature_df['INSTANCE_NAME'] = feature_df.index
feature_df = feature_df.reset_index()
feature_df = feature_df.drop(columns=['index'])
feature_df.head()
```

```python
for i in feature_df.columns:
    try:
        feature_df[[i]] = feature_df[[i]].astype(float)
    except:
        pass
```

### Data Analysis

```python
import pandas as pd
import plotly.graph_objs as go
import dash
import dash_core_components as dcc
import dash_html_components as html
from jupyter_dash import JupyterDash

app = JupyterDash(__name__)

# load the dataframe
df = feature_df

# create a dropdown menu with the column names
column_options = [{"label": col, "value": col} for col in df.columns]
dropdown = dcc.Dropdown(
    id="column-dropdown",
    options=column_options,
    value=[df.columns[0]],
    multi=True,
    style={"display": "block", "margin-left": "auto", "margin-right": "auto", "width": "50%"},
)

# create the plot
plot = dcc.Graph(id="boxplot")

app.layout = html.Div([dropdown, plot])

@app.callback(
    dash.dependencies.Output("boxplot", "figure"),
    [dash.dependencies.Input("column-dropdown", "value")],
)
def update_plot(column_names):
    box_data = [
        go.Box(
            x=df[column_name],
            name=column_name,
            boxmean="sd",
            boxpoints="all",
            jitter=0.3,
            pointpos=-1.8,
        )
        for column_name in column_names
    ]
    scatter_data = [
        go.Scatter(
            x=df[column_name],
            y=[column_name] * len(df),
            mode="markers",
            name=column_name,
            text=[f"file: {feature_df['INSTANCE_NAME'][index]} value: {x:.2f}" for index, x in enumerate(df[column_name])],
            #text=feature_df['lp_file'],
            hoverinfo="text",
            showlegend=False,
        )
        for column_name in column_names
    ]
    data = box_data + scatter_data
    layout = go.Layout(
        xaxis={"title": "Value"},
        yaxis={"title": "Variable"},
        title="Boxplots for the SCIP MIP solver's features",
    )
    fig = go.Figure(data=data, layout=layout)
    return fig

app.run_server(mode="inline")
```

```python
# solved by one and not the other

def compare_solved(A, B, n_instances, timeout):
    '''
    Input: 
        - A: represents one method
        - B: represents the other method
        - n_instances: number of instances
        - timeout: Value of timeout (note if timeout is par10ified then include this as timeout e.g. par10 of 9600 is 96000)
    
    Output:
        - countA: Number of instances solved by A but not B
        - countB: Number of instances solved by B but not A
        - count0: Number of instances solved by neither method

    Solved: Solving in this case is does it reach the timeout or not?

    '''
    solvedbya = []
    notsolvedbya = []
    solvedbyb = []
    notsolvedbyb = []

    for i in range(n_instances):
        # selector A
        if A[i] == timeout:
            notsolvedbya.append(i)
        else:
            solvedbya.append(i)
        # selector B
        if B[i] == timeout:
            notsolvedbyb.append(i)
        else:
            solvedbyb.append(i)

    A_solved = set(solvedbya).difference(set(solvedbyb))
    B_solved = set(solvedbyb).difference(set(solvedbya))
    neither = set(notsolvedbya).intersection(set(notsolvedbyb))
    countA = len(A_solved)
    countB = len(B_solved)
    count0 = len(neither)
    return countA, countB, count0


compare_solved(A=df.autofolio, B=df.sbs, n_instances=10000, timeout=96000)
```

```python
def compare_times(A, B, n_instances):
    '''
    Input: 
        - A: represents one method
        - B: represents the other method
        - n_instances: number of instances
    
    Output:
        - countA: Number of instances solved faster by A 
        - countB: Number of instances solved faster by B 
        - count0: Number of instances solved by both equally fast

    Solved: Solving in this case is which method took less time?

    '''
    solvedbya = []
    solvedbyb = []
    equal = []

    for i in range(n_instances):
        if A[i] > B[i]:
            solvedbyb.append(i)
        elif A[i] < B[i]:
            solvedbya.append(i)
        else:
            equal.append(i)

    countA = len(solvedbya)
    countB = len(solvedbyb)
    count0 = len(equal)
    return countA, countB, count0


compare_times(A=df.autofolio, B=df.sbs, n_instances=10000)
```
